```python
def GranularSemanticFilling(Data):                             # Data filling with granular semantics
  import heapq
  from sklearn.model_selection import train_test_split
  from sklearn.linear_model import LinearRegression
 #print(matrix.shape)
  Xa=np.array(Data)
  XAr=np.zeros(Xa.shape)
  XAr[:,:]=Xa[:,:]
  ps1 = np.where(Xa==0.007)
  rows=ps1[0]
  cols=ps1[1]
  X_Del_D=np.zeros((Xa.shape))
  X_Del_D[:,:]=Xa[:,:]
  X_Del_D=np.delete(X_Del_D,rows,0)
  dframe=pd.DataFrame(X_Del_D, columns=None)
  matrix=dframe.corr()
  matrixar=np.array(matrix.values)
  matrix=dframe.corr()                                                # finding feature correlation
  matrixar=np.array(matrix.values)
  regressor = LinearRegression()
  dhm=matrixar.shape
  mat=np.zeros((dhm[0]-1,1))
  wsize=5
  matp=np.zeros((wsize,wsize))
  count=0
  for icr in range(len(ps1[0])):
 # print(icr)
     rr=rows[icr]
     rc=cols[icr]
     rmn=rr-wsize+1
     matp=np.zeros((wsize,wsize))
     mat=np.zeros((dhm[0]))
     mat[1:64]=matrixar[1:64,rc]
     simpos1=heapq.nlargest(wsize+1, range(len(mat)), mat.take)                  # finding the top 'wsize' number of similar features to that of the query feature
     simpos=np.delete(simpos1,0,0)
     tempMat=np.zeros((100,wsize))
     if rmn>=0:                                                                 #######################
        matp[0:wsize,0:wsize]=XAr[1:wsize+1,simpos1[0:wsize]]                   ####################### Formation of Granules
        #matp[0:wsize,wsize-1]=XAr[rmn:rr+1,rc]
        matp[:,[0,wsize-1]]=matp[:,[wsize-1,0]]
        matp[wsize-1,wsize-1]=0.5
        psmt=np.where(matp==0.007)
        szmt=len(psmt[0])
        if szmt==0:
           matp=matp
        else:
           rcmt=ps[0]
           for icmt in range(szmt-1):
               matp[rcmt[icmt],:]=X_Del_D[icmt,:]                                  ######################
        Xt=matp[0:wsize,0:wsize-1]
        yt=matp[0:wsize,wsize-1]
     else:
        rnmx=rows[icr]+wsize-1
        matp[:,:]=XAr[rr:rnmx+1,simpos1[0:5]]
        matp[0,wsize-1]=matp[wsize-1,0]
        Xt=matp[0:wsize,0:wsize]
        yt=matp[0:wsize,wsize-1]
     X_train, X_test, y_train, y_test = train_test_split(Xt,yt.transpose(),test_size=1/wsize,shuffle=False)
     regressor.fit(X_train,y_train)
     test_d=matp[wsize-1,0:wsize-1]
     y_pred = regressor.predict(X_test)
     XAr[rows[icr],cols[icr]]=y_pred                                                     # Granular predicition of Missing Data and filling it
     count=count+1
  return(XAr, matp)

[XAr, matp]=GranularSemanticFilling(Data)
#print(simpos)
#print(matp)
#print(Xa[rmn:rr+1,rc])
```

```python
from collections import Counter
from imblearn.over_sampling import SMOTE

# x = df.iloc[:,:-1]
x = dframeF.iloc[:,[0,20,24,26,33,34,36,37,44,45,49,52,55,57,58,59]]
y = dframeF['Labels']

'''Increase the bankcruptcy samples from 2049 to 40000'''

sm = SMOTE(random_state=42,sampling_strategy={1:40000})
X_res, y_res = sm.fit_resample(x, y)
print('Resampled dataset shape {}'.format(Counter(y_res)))
print(X_res.shape)
print(y_res.shape)
```

```python
'''Split the resampled data into train & test data with 70:30 mix'''

xtrain, xtest, ytrain, ytest = train_test_split(X_res, y_res, test_size=0.30, random_state=0)
print('xtrain shape')
print(xtrain.shape)
print('xtest shape')
print(xtest.shape)
```

```python
'''Random Forest Classifier on selected features data'''

from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics

rfmodel = RandomForestClassifier()
rfmodel.fit(xtrain,ytrain)
print('model')
print(rfmodel)

'''Training accuracy'''
ytrainpred=rfmodel.predict(xtrain)
print('confusion matrix')
print(metrics.confusion_matrix(ytrain, ytrainpred))
print('classification report')
print(metrics.classification_report(ytrain, ytrainpred))
print('Accuracy : %f' % (metrics.accuracy_score(ytrain, ytrainpred)))
print('Area under the curve : %f' % (metrics.roc_auc_score(ytrain, ytrainpred)))

'''Testing accuracy'''
ypredrf = rfmodel.predict(xtest)
print('\n\n\nconfusion matrix')
print(metrics.confusion_matrix(ytest, ypredrf))
print('classification report')
print(metrics.classification_report(ytest, ypredrf))
print('Accuracy : %f' % (metrics.accuracy_score(ytest, ypredrf)))
print('Area under the curve : %f' % (metrics.roc_auc_score(ytest, ypredrf)))
```

#### We can see an increase in F1 score and the Area under the curve. This is the advantage of synthetic over sampling technique which has given a good lift in accuracy.
